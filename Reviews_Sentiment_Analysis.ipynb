Scraping Google Play Reviews of an app and performing sentiment analysis using transformers(Sentiment). 

I have used the online fashion app MANGO!

```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all;#Scraping data
```

```python
#adding the app id and other details
mango=reviews_all('com.dylvian.mango.activities',sleep_milliseconds=0,lang='en',sort=Sort.NEWEST)
```

```python
mango
```

```python
df=pd.json_normalize(mango)
```

```python
df.head(500)
```

```python
df['score'].mean()
```

```python
df['reviewCreatedVersion'].value_counts()
```

```python
from transformers import pipeline
#using a pre-trained model from Huggging Face
sentiment_analysis=pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```

```python
print(sentiment_analysis("I Loved it!!"))
```

```python
df.dtypes
```

```python
df['content']=df['content'].astype('str')
```

```python
#creating a new column to store the sentiment and the score
df['result']=df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
df.head(500)
```

```python
y=sentiment_analysis("I Loved it!!")
```

```python
y[0]
```

```python
df['sentiment']=df['result'].apply(lambda x: (x[0]['label']))
df['score']=df['result'].apply(lambda x: (x[0]['score']))
```

```python
df.head(500)
```

```python
df['score'].mean()
```
